```python
# read  
webget_2016 = pd.read_csv("../dataset/webget_2016.csv", sep = ',', encoding = "ISO-8859-1")
webget_df_2016 = webget_2016[webget_2016['dtime'] <= '2017-07-01']
webget_df_2016 = webget_df_2016[webget_df_2016['ttfb_avg'] > 0]
```
