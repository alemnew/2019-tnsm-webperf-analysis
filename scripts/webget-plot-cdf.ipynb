```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
import requests 
from imp import reload 
import Util; reload(Util)
import Cdf
import Pmf
from matplotlib.ticker import FuncFormatter
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv(Util.dataset, sep = ',', encoding = "ISO-8859-1")
```

```python
data = data[data['lookup_avg_time'] <=200000]
data = data[data['ttfb_avg'] <=1000000]
data = data[data['ttfb_avg'] >=10000]
data = data[data['fetch_time'] >=100000]
data.count()
```

```python
data[['target', 'objects', 'threads', 'requests', 'connections', 'reused_connections']].tail(15)
```

```python
# data prepareation 
data['dtime'] = data['dtime'].apply(Util.round_time_to_day)
data = data[data['dtime'] >= '2014-01-01']
```

```python
data.columns
```

```python
websites = Util.tables_site(data)
ggl_fetch_time = []
ggl_avg_lookup_time = []
ggl_avg_ttfb = []
ggl_objects = []
ggl_bytes_tot = []

fb_fetch_time = []
fb_avg_lookup_time = []
fb_avg_ttfb = []
fb_objects = []
fb_bytes_tot = []

ut_fetch_time = []
ut_avg_lookup_time = []
ut_avg_ttfb = []
ut_objects = []
ut_bytes_tot = []

for i in pd.unique(data['target']):
    if "google.com" in i:
        ggl_fetch_time = websites[i]['fetch_time'] / 1000
        ggl_avg_lookup_time = websites[i]['lookup_avg_time'] / 1000
        ggl_avg_ttfb = websites[i]['ttfb_avg'] / 1000
        ggl_objects = websites[i]['objects']
        ggl_bytes_tot = websites[i]['bytes_total'] / 1000
    elif "facebook.com" in i:
        fb_fetch_time = websites[i]['fetch_time'] / 1000
        fb_avg_lookup_time = websites[i]['lookup_avg_time'] / 1000
        fb_avg_ttfb = websites[i]['ttfb_avg'] / 1000
        fb_objects = websites[i]['objects']
        fb_bytes_tot = websites[i]['bytes_total'] / 1000
    elif "youtube.com" in i:
        ut_fetch_time = websites[i]['fetch_time'] / 1000
        ut_avg_lookup_time = websites[i]['lookup_avg_time'] / 1000
        ut_avg_ttfb = websites[i]['ttfb_avg'] / 1000 
        ut_objects = websites[i]['objects']
        ut_bytes_tot = websites[i]['bytes_total'] / 1000
        
thrput = data['bytes_sec'].div(100000)
```

```python
cgft = Cdf.MakeCdfFromList(ggl_fetch_time)
cglt = Cdf.MakeCdfFromList(ggl_avg_lookup_time)
cgtt = Cdf.MakeCdfFromList(ggl_avg_ttfb)
cgo = Cdf.MakeCdfFromList(ggl_objects)
cgbt = Cdf.MakeCdfFromList(ggl_bytes_tot)

cfft = Cdf.MakeCdfFromList(fb_fetch_time)
cflt = Cdf.MakeCdfFromList(fb_avg_lookup_time)
cftt = Cdf.MakeCdfFromList(fb_avg_ttfb)
cfo = Cdf.MakeCdfFromList(fb_objects)
cfbt = Cdf.MakeCdfFromList(fb_bytes_tot)

cuft = Cdf.MakeCdfFromList(ut_fetch_time)
cult = Cdf.MakeCdfFromList(ut_avg_lookup_time)
cutt = Cdf.MakeCdfFromList(ut_avg_ttfb)
cuo = Cdf.MakeCdfFromList(ut_objects)
cubt = Cdf.MakeCdfFromList(ut_bytes_tot)

cthrput = Cdf.MakeCdfFromList(thrput)
```
