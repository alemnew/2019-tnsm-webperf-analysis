```python
# imports and env't configs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import json
import requests 
from imp import reload 
import Util; reload(Util)
from datetime import datetime as dt
import matplotlib.dates as md
import warnings
import seaborn as sns
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
# data reading 
data = pd.read_csv(Util.dataset, sep = ',', encoding = "ISO-8859-1")
```

```python
# data prepareation 
data['dtime'] = data['dtime'].apply(Util.round_time_to_month)
data = data[(data['dtime'] >= '2014-01') & (data['dtime'] < '2017-07')]
```

```python
data['fetch_time'] = data['fetch_time'].div(1000)
data['ttfb_avg'] = data['ttfb_avg'].div(1000)
data['lookup_avg_time'] = data['lookup_avg_time'].div(1000)
```

```python
# Filter the dataset for each website ...
google = data[data['target'].str.contains("google.com")]
facebook = data[data['target'].str.contains("facebook.com")]
youtube = data[data['target'].str.contains("youtube.com")]
```

```python
# Group data first by probe ID, then by dtime (time)
google_grouped = google.groupby(["unit_id", "dtime"])
facebook_grouped = facebook.groupby(["unit_id", "dtime"])
youtube_grouped = youtube.groupby(["unit_id", "dtime"])
```

```python
# For each group, calculate the average of only the snow_depth column
google_ft_mnth_median = google_grouped.aggregate({"fetch_time":np.median})
google_ttfb_mnth_median = google_grouped.aggregate({"ttfb_avg":np.median})
google_dnst_mnth_median = google_grouped.aggregate({"lookup_avg_time":np.median})
google_objects_mnth_median = google_grouped.aggregate({"objects":np.median})
google_bytes_mnth_median = google_grouped.aggregate({"bytes_total":np.median})

facebook_ft_mnth_median = facebook_grouped.aggregate({"fetch_time":np.median})
facebook_ttfb_mnth_median = facebook_grouped.aggregate({"ttfb_avg":np.median})
facebook_dnst_mnth_median = facebook_grouped.aggregate({"lookup_avg_time":np.median})
facebook_objects_mnth_median = facebook_grouped.aggregate({"objects":np.median})
facebook_bytes_mnth_median = facebook_grouped.aggregate({"bytes_total":np.median})

youtube_ft_mnth_median = youtube_grouped.aggregate({"fetch_time":np.median})
youtube_ttfb_mnth_median = youtube_grouped.aggregate({"ttfb_avg":np.median})
youtube_dnst_mnth_median = youtube_grouped.aggregate({"lookup_avg_time":np.median})
youtube_objects_mnth_median = youtube_grouped.aggregate({"objects":np.median})
youtube_bytes_mnth_median = youtube_grouped.aggregate({"bytes_total":np.median})
```

```python
ggl_df = google_ttfb_mnth_median.reset_index(level=['unit_id', 'dtime'])
ggl_df['fetch_time'] = google_ft_mnth_median.reset_index(level=['unit_id', 'dtime'])['fetch_time']
ggl_df['lookup_avg_time'] = google_dnst_mnth_median.reset_index(level=['unit_id', 'dtime'])['lookup_avg_time']
ggl_df['objects'] = google_objects_mnth_median.reset_index(level=['unit_id', 'dtime'])['objects']
ggl_df['bytes_total'] = google_bytes_mnth_median.reset_index(level=['unit_id', 'dtime'])['bytes_total']

fb_df = facebook_ttfb_mnth_median.reset_index(level=['unit_id', 'dtime'])
fb_df['fetch_time'] = facebook_ft_mnth_median.reset_index(level=['unit_id', 'dtime'])['fetch_time']
fb_df['lookup_avg_time'] = facebook_dnst_mnth_median.reset_index(level=['unit_id', 'dtime'])['lookup_avg_time']
fb_df['objects'] = facebook_objects_mnth_median.reset_index(level=['unit_id', 'dtime'])['objects']
fb_df['bytes_total'] = facebook_bytes_mnth_median.reset_index(level=['unit_id', 'dtime'])['bytes_total']

ut_df = youtube_ttfb_mnth_median.reset_index(level=['unit_id', 'dtime'])
ut_df['fetch_time'] = youtube_ft_mnth_median.reset_index(level=['unit_id', 'dtime'])['fetch_time']
ut_df['lookup_avg_time'] = youtube_dnst_mnth_median.reset_index(level=['unit_id', 'dtime'])['lookup_avg_time']
ut_df['objects'] = youtube_objects_mnth_median.reset_index(level=['unit_id', 'dtime'])['objects']
ut_df['bytes_total'] = youtube_bytes_mnth_median.reset_index(level=['unit_id', 'dtime'])['bytes_total']
```

```python
ggl_df['bytes_total'] = ggl_df['bytes_total'].div(1000)
fb_df['bytes_total'] = fb_df['bytes_total'].div(1000)
ut_df['bytes_total'] = ut_df['bytes_total'].div(1000)
```

```python
def stop_plot_box(ax, region, shrink_right = False, fs = 18, plt_type = 'fetch_type'):
    ax.set_title("")
    plt.suptitle("")
    ax.grid(False)
    
    # iterate over boxes and change the colors
    for i,box in enumerate(ax.artists):
        box.set_edgecolor('black')
        box.set_facecolor('white')

    if plt_type == 'fetch_time': 
        ax.set_ylim([100, 10000])
        ax.set_yscale('log')
    elif plt_type == 'ttfb_avg':
        ax.set_ylim([10, 1000])
        ax.set_yscale('log')
    elif plt_type == 'dns_time':
        ax.set_ylim([1, 501])
        ax.set_yscale('log')
    elif plt_type == 'obj':
        ax.set_ylim([1, 101])
        ax.set_yscale('log')
    elif plt_type == 'byt':
        ax.set_ylim([10, 10000])
        ax.set_yscale('log')

    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.spines['bottom'].set_position(('axes', -0.05))
    ax.spines['left'].set_position(('axes', -0.015))
    ax.set_xlabel('', fontsize=fs)
    ax.set_title(region,  fontsize=18)
    major_ticks = np.arange(0, 42, 6)                                              
    minor_ticks = np.arange(0, 42, 1) 
    ax.set_xticks(minor_ticks, minor=True)                                           
    ax.set_xticks(major_ticks, minor=False)                                           

    labels = ['Jan\n2014 ','Jul', 'Jan\n2015 ','Jul',
          'Jan\n2016 ',' Jul','Jan\n2017 ']
    ax.set_xticklabels(labels)  
    ax.set_ylabel('', fontsize=10)

def boxplot_sorted(df, by, column, ax, rot = 0):
    medianprops = dict(linestyle='--', color = 'blue', linewidth=2)
    props = dict(boxes="black", whiskers="black", medians="black") #for colors -- change this for color scale
    df2 = pd.DataFrame({col:vals[column] for col, vals in df.groupby(by)})
    meds = df2.median()#.sort_values() # for sorting by median 
         
    return df2[meds.index].plot.box(ax = ax, rot = rot,  vert = True, color = props,
                                   medianprops = medianprops, showfliers = False, 
                                   return_type = "axes")
```

```python
fig = plt.figure(figsize=(8,4))
plt.rcParams['font.size'] = 12

ax1 = fig.add_subplot(211)
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 1})
g = sns.pointplot(x='dtime', y='lookup_avg_time', 
              data=fb_df.groupby('dtime', as_index=False).median(), ax=ax1,
             markers=["*"], linestyles=["-"], color="coral",
             join=True)
g.set(xticklabels=[], yticklabels=[])

ax1 = sns.boxplot(x="dtime", y="lookup_avg_time", data=fb_df,linewidth=1, width=0.6, showfliers=False, ax=ax1)
stop_plot_box(ax1, region='Google', fs = 18, plt_type = 'dns_time')

ax1.legend(["Median"], bbox_to_anchor=(0.5, 0.7), frameon=False)
ax1.set_ylabel('DNS Lookup Time (ms)', fontsize=10)
```
