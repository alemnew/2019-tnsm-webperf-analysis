```python
# imports and env't config 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.gridspec as gridspec
import json
import requests 
from imp import reload 
import Util; reload(Util)
from pygeocoder import Geocoder as gc
import warnings
# from pygeolib import GeocoderError
import time 
import Cdf
import Pmf
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
# read  
webget_2016 = pd.read_csv("../dataset/webget_2016.csv", sep = ',', encoding = "ISO-8859-1")
webget_df_2016 = webget_2016[webget_2016['dtime'] <= '2017-07-01']
webget_df_2016 = webget_df_2016[webget_df_2016['ttfb_avg'] > 0]
webget_df_2016 = webget_df_2016[webget_df_2016['fetch_time'] > 100000]
```

```python
webget_df_2016 = webget_df_2016[webget_2016['country'] != '']
webget_df_2016 = webget_df_2016.dropna(subset=['country'])
webget_df_2016['fetch_time'] = webget_df_2016['fetch_time'].div(1000)
webget_df_2016['ttfb_avg'] = webget_df_2016['ttfb_avg'].div(1000)
webget_df_2016['lookup_avg_time'] = webget_df_2016['lookup_avg_time'].div(1000)

google_16 = webget_df_2016[webget_df_2016['target'].str.contains("google.com")]
facebook_16 = webget_df_2016[webget_df_2016['target'].str.contains("facebook.com")]
youtube_16 = webget_df_2016[webget_df_2016['target'].str.contains("youtube.com")]
```

```python
eu_1 = ['Belgium', 'France', 'Italy', 'Spain', 'Germany', 'Romania', 'Latvia', \
        'Netherlands' , 'Switzerland', 'Czechia', 'Finland', 'Norway', 'Sweden', \
        'United Kingdom', 'Ireland', 'Slovenia','Israel', 'Bosnia and Herzegovina']
# eu_2 = ['Finland', 'Norway', 'Sweden', 'United Kingdom', 'Ireland', 'Slovenia']
us_ca = ['United States', 'Canada']
others = ['South Africa', 'Philippines', 'Singapore', 'New Zealand', 'Japan', \
          'Brazil', 'Australia']
```
